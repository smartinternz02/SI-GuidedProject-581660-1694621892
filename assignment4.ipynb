## Step 5: Splitting into dependent and independent variables

```python
x=df.copy()
x.drop(['StandardHours','EmployeeCount','Over18','Attrition'],axis=1,inplace=True)
y = df['Attrition']
```

```python
x.head()
```

## Step 6: Encoding

```python
# Performing one-hot encoding on categorical columns
x_encoded=pd.get_dummies(x, drop_first=True)
x_encoded.head()
```

## Step 7: Feature scaling 

```python
from sklearn.preprocessing import MinMaxScaler

ms= MinMaxScaler()
x_scaled=pd.DataFrame(ms.fit_transform(x_encoded),columns=x_encoded.columns)
x_scaled
```

## Step 8:Splitting data into train and test 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Model building

## Step 1: Importing model building libraries

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

## Step 2: Initialising the model

```python
logistic_model = LogisticRegression(random_state=42)
decision_tree_model = DecisionTreeClassifier(random_state=42)
random_forest_model = RandomForestClassifier(random_state=42)
```

## Step 3: Training and testing model

```python
# Training and testing the Logistic Regression model
logistic_model.fit(x_train, y_train)
logistic_pred = logistic_model.predict(x_test)

# Training and testing the Decision Tree model
decision_tree_model.fit(x_train, y_train)
decision_tree_pred = decision_tree_model.predict(x_test)

# Training and testing the Random Forest model
random_forest_model.fit(x_train, y_train)
random_forest_pred = random_forest_model.predict(x_test)
```
